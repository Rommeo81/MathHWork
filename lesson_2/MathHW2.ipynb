```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# зададим область для отрисовки графики от 0 до 10 по линии X и количество точек, равное 2000
x=np.linspace(0, 10, 2000)

# зададим размер рисунка отображаемого графика
plt.figure(figsize=(10,10))

# зададим функцию параболы
y = (-17/8)*(x**2)+(25/2)*x-(67/8)

# зададим точки, через которые должнен проходить функция
x1, y1 = 1, 2
x2, y2 = 3, 10
x3, y3 = 5, 1

# зададим параметры длинны осей на графике (x от 0 до 6, y от -2 до 12)
plt.axis([0, 6, -2, 12])

# отрисуем сам график функции
plt.plot(x, y, c="g")

# добавим отображение точек на график
plt.plot(x1, y1, c="b", marker='o')
plt.plot(x2, y2, c="b", marker='o')
plt.plot(x3, y3, c="b", marker='o')

# добавим название графика
plt.title("График параболы (-17/8)*(x**2)+(25/2)*x-(67/8) с точками (1, 2), (3, 10), (5, 1)")

# добавим подписи осей
plt.xlabel('Ось x')
plt.ylabel('Ось y')

# добавим отображение сетки на графике
plt.grid()

# выводим все на экран
plt.show()
```
