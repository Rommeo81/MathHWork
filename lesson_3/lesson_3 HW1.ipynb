```python
# Задание 1. Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
# Найдите их сумму. (на листочке). Смотреть фото, прикрепленное к ДЗ.
# Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


# создаём функцию "SumVect" выполняющую сложение векторов
def SumVect(vect1, vect2):
    return [i + j for i, j in zip(vect1, vect2)]

# задаём значение дувух векторов
vect1 = [10, 10, 10]
vect2 = [0, 0, -10]

# запускаем функцию и выводим все на экран
print(f"Сумма векторов {vect1} и {vect2} будет {SumVect(vect1, vect2)}")
```

```python
# Доказательство перпендикулярности прямых смотреть на фото, прикрепленное к ДЗ.
# Прямые не выглядят перпендикулярно из-за не заданной координатной сетки и из=за не заданного размера отображаемого рисунка
# Для наглядного отображения перпендикулярности прямых допишем пару строк:

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-5, 5, 21)

# зададим размер отображаемого графика с одинаковой длиной сторон 5 на 5
plt.figure(figsize=(5,5))

y = 3*x+1
y2 = (-1/3)*x+1

# добавим одинаковую длинну значений осей X и Y
plt.axis([-5, 5, -5, 5])

plt.plot(x,y)
plt.plot(x,y2)

plt.xlabel("x")
plt.ylabel("y")

# добавим отрисовку линии сетки
plt.grid()
```

```python
# Задание 3.1 (в программе)
# Напишите код на Python, реализующий построение графика окружности.

# Значение каждой точки будем расчитывать в каждой четверти графика беря значения X и -X для каждой Y и -Y

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math

x1 = []
y1 = []
x2 = []
y2 = []

# т.к. значения индекса начинается с 0, производить расчет будет на 1 раз больше, чем заданный радиус,
# что бы рассчитываемы сегмент отобразажлся полноценно
for i in range(1001):
    
    # зададим радиус окружности
    r = 1000
    
    # добавим значения X и -X
    x1.append(i)
    x2.append(-i)
    
    # добавим расчет каждого значения Y и -Y
    y1.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))


plt.figure(figsize = (7, 7))

# отобразим верхний правый отрезок окружности
plt.plot(x1, y1, color='r')
# отобразим нижний правый отрезок окружности
plt.plot(x1, y2, color='r')
# отобразим верхний левый отрезок окружности
plt.plot(x2, y1, color='r')
# отобразим нижний левый отрезок окружности
plt.plot(x2, y2, color='r')

plt.axis('scaled')

plt.grid()

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# Задание 3.2 (в программе)
# Напишите код на Python, реализующий построение графика эллипса.

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x1 = []
x2 = []
y1 = []
y2 = []

for i in range(401):
    a = 400
    b = 200
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))

plt.figure(figsize = (7, 10))    

plt.plot(x1, y1, color='r')
plt.plot(x1, y2, color='r')
plt.plot(x2, y1, color='r')
plt.plot(x2, y2, color='r')

plt.axis('scaled')

plt.grid()

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
