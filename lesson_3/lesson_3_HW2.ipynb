```python
# Задание 1 (в программе)
# Нарисуйте график функции: для некоторых (2-3 различных) значений параметров k, a, b

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-2*np.pi, 3*np.pi, 1000)

k1, a1, b1 = np.linspace(5, 10, num = 3)
k2, a2, b2 = np.linspace(0.7, 20, num = 3)
k3, a3, b3 = np.linspace(9, 30, num = 3)

print(f'k1={k1}, a1={a1}, b1={b1}')
print(f'k2={k2}, a2={a2}, b2={b2}')
print(f'k3={k3}, a3={a3}, b3={b3}')

plt.figure(figsize = (10, 5))

plt.plot(x, k1 * np.cos(x - a1) + b1, color='r', label='k1, a1, b1')
plt.plot(x, k2 * np.sin(x - a2) + b2, color='g', label='k2, a2, b2')
plt.plot(x, k3 * np.cos(x - a3) + b2, color='b', label='k3, a3, b3')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.legend(frameon=True)

plt.show()
```

```python
# Задание 4.1 (в программе)
# Решите систему уравнений:
# exp(x) + x∙(1 – y) = 1  
# y = x2 – 1

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-3, 3, 1000)

plt.figure(figsize = (5, 10))

plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)

plt.xlabel('x')
plt.ylabel('y')

plt.ylim(-1,5)

plt.grid(True)

plt.axis('scaled')

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
            
print (f'Точка пересечения А по координатам ({round(x1,3)}, {round(y1,3)})')
print (f'Точка пересечения B по координатам ({round(x2,3)}, {round(y2,3)})')
```

```python
# Задание 4.2 (в программе)
# Решите систему неравенств:

# y = x2 – 1
# exp(x) + x∙(1 – y)  - 1 > 0

x = np.linspace(-2, 3, 201)

plt.figure(figsize = (5, 10))

plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (f'Точка пересечения А по координатам ({x1}, {y1})')
print (f'Точка пересечения B по координатам ({x2}, {y2})')
print (f'Точка пересечения С по координатам ({x3}, {y3})')
```
